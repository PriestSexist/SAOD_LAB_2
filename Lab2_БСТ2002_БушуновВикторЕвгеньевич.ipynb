```python
class Node:
    # конструктор и мы ему передаём х
    def __init__(self, x):
        self.left = None
        self.right = None
        self.data = x


class Tree:
    root = None

    def insert(self, x):
        # если пусто, то записываем
        if self.root is None:
            self.root = Node(x)
            return
        # если меньше, то идём влево
        if x < self.root.data:
            if self.root.left is None:
                self.root.left = Tree()
            self.root.left.insert(x)
        # если больше, то идём вправо
        else:
            if self.root.right is None:
                self.root.right = Tree()
            self.root.right.insert(x)

    def find(self, x):
        if self.root is None:
            return None
        if x == self.root.data:
            return self.root
        if x < self.root.data:
            if self.root.left is None:
                return None
            return self.root.left.find(x)
        else:
            if self.root.right is None:
                return None
            return self.root.right.find(x)


T = Tree()
n = int(input("Введите количество чисел: "))

for i in range(n):
    m = int(input("Введите число: "))
    T.insert(m)
x = int(input("Введитие число для поиска: "))
print(T.find(x))
```

```python
import time

def interpol_search(a, x):
    low = 0
    high = len(a) - 1
    while low <= high:
        # Вычисление среднего значения (основной код алгоритма интерполяции)
        mid = low + int((high - low) * (x - a[low])/(a[high] - a[low]))
        if x < a[mid]:
            high = mid - 1
        elif x > a[mid]:
            low = mid + 1
        else:
            return mid


def insert(a, x):
    start = 0
    end = len(a)

    while (start <= end):
        mid = (start + end) // 2
        if x == a[mid]:
            break
        elif x < a[mid]:
            end = mid - 1
        else:
            start = mid + 1

    b = a[:mid]
    b.append(x)
    b += a[mid:]
    return b

entered = input("Введите список чисел, разделенных пробелом: ").split()

a = list(map(int, entered))
x = int(input("Введите число для поиска: "))
y = int(input("Введите число для вставки: "))
z = int(input("Введите число для удаления: "))

start_time = time.time()

i = interpol_search(a, x)
b=insert(a, y)

end_time = time.time() - start_time

print("Индекс найденного элемента: ", i)
print("Изменённая строка после добавления: ", b)
a.remove(z)
print("Изменённая строка после удаления: ", a)
print("Время выполнения: ", '%.9f'%end_time, "sec")
```

```python
class HMap:
    def __init__(self):
        self.size = 11
        self.index = [[] for i in range(self.size)]

    def hash_f(self, key, size):
        return key % size

    def insert(self, val):
        self.index[self.hash_f(val, self.size)].append(val)
        return

    def search(self, val):
        for i in range(len(self.index[self.hash_f(val, self.size)])):
            if self.index[self.hash_f(val, self.size)][i] == val:
                return i
        return None

    def delete(self, val):
        for i in range(len(self.index[self.hash_f(val, self.size)])):
            if self.index[self.hash_f(val, self.size)][i] == val:
                self.index[self.hash_f(val, self.size)][:] = self.index[self.hash_f(val, self.size)][:i] + self.index[self.hash_f(val, self.size)][i+1:]
                return self.index[self.hash_f(val, self.size)]

    def print(self):
        for i in self.index:
            print(i)


hmap = HMap()
hmap.insert(5)
hmap.insert(16)
hmap.insert(7)
hmap.insert(27)
hmap.insert(38)
hmap.insert(8)
print(hmap.search(7))
hmap.delete(16)
print("====================")
hmap.print()
```
